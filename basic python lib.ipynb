```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df1=pd.read_csv('Universities.csv')
```

```python
df1
```

```python
plt.hist(df1['SAT'])
plt.xlabel('SATSCORES')
plt.ylabel('Frequency')
plt.title('venky')
plt.show()
```

```python
plt.hist(df1)
```

```python
sns.histplot(df1['SAT'], color='green')
plt.xlabel('sat scores')
plt.ylabel('frequency')
plt.title('venky')
plt.show()
```

```python
plt.plot(df1['SAT'],df1['Accept'],'ro')
plt.xlabel('sat scores')
plt.ylabel('Acceptance Rate')
plt.title('SAT Scores vs. Acceptance Rate')
plt.show()
```

```python
plt.scatter(df1['SAT'], df1['Accept'])
plt.xlabel('SAT Scores')
plt.ylabel('Acceptance Rate')
plt.title('SAT Scores vs. Acceptance Rate')
plt.show()
```

```python
sns.pairplot(df1)
plt.show()
```

```python
plt.bar(df1.index, df1['SAT'])
plt.xlabel('Index')
plt.ylabel('SAT Scores')
plt.title('SAT Scores by Index')
plt.show()
```
